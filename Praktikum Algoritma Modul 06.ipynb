```python
#Farhan Raditya
#065002500021

def konversinilai(grade, default=None):
    grades_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
    return grades_map.get(grade.strip().upper(), default)

total = 0
jumlah_nilai = 0

while True:
    nilai_huruf = input("masukkan nilai: ")

    if nilai_huruf == "":
        break

    nilai_angka = konversinilai(nilai_huruf)

    if nilai_angka is not None:
        print(f"nilai = {nilai_angka}")
        total += nilai_angka
        jumlah_nilai += 1
    else:
        print("nilai = 0")

if jumlah_nilai > 0:
    rata_rata = total / jumlah_nilai
    print(f"Rata-ratanya adalah: {rata_rata}")
else:
    print("Tidak ada nilai yang dimasukkan.")
```
