```python
#Farhan Raditya
#065002500021
#LATIHAN PERTAMA

def hitung_nilai_rata_rata():
    """
    menghitung nilai rata-rata berdasarkan input nilai huruf dari pengguna.
    """
    # buat nilai rata-rata huruf ke angka
    peta_nilai = {
        "A": 4.00,
        "A-": 3.75,
        "B+": 3.50,
        "B": 3.00,
        "B-": 2.75,
        "C+": 2.50,
        "C": 2.00,
        "C-": 1.75,
        "D": 1.50,
        "E": 1.25
    }

    # menginput nilai dari user
    print("tekan enter untuk selesai dan menghitung rata-rata.")

    # inisialisasi daftar nilai
    daftar_nilai = []

    # memulai program loop input
    while True:
        nilai_huruf = input("Masukkan nilai: ")

        # memeriksa kondisi berhenti
        if nilai_huruf == "":
            break # Keluar dari loop jika input kosong

        # memvalidasi input
        if nilai_huruf in peta_nilai:
            # Dapatkan nilai numerik dari dictionary
            nilai_angka = peta_nilai[nilai_huruf]
            # Tambahkan nilai numerik ke dalam list
            daftar_nilai.append(nilai_angka)
            print(f"Nilai {nilai_huruf} (Setara: {nilai_angka}) ditambahkan.")
        else:
            # menangani input yang tidak valid
            print(f"Error: '{nilai_huruf}' ini bukan nilai yang valid, silakan coba lagi.")

    # menghitung rata-rata
    if len(daftar_nilai) > 0:
        total_nilai = sum(daftar_nilai)
        jumlah_nilai = len(daftar_nilai)
        rata_rata = total_nilai / jumlah_nilai

        # cetak hasilnya
        print("\n--- Perhitungan Selesai ---")
        print(f"jumlah nilai yang dimasukkan: {jumlah_nilai}")
        print(f"total poin: {total_nilai}")
        print(f"nilai rata-rata anda adalah: {rata_rata:.2f}")
    else:
        # 8. Tangani jika tidak ada nilai yang dimasukkan
        print("\ntidak ada nilai valid yang dimasukkan.")

hitung_nilai_rata_rata()
```

```python
#Farhan Raditya
#065002500021
#LATIHAN kEDUA

total_harga = 0.0
print("Program Hitung Tiket Kebun Binatang")
print("Masukkan umur (kosongkan untuk bayar):")

# Memulai perulangan (loop)
while True:
    umur_input = input("masukkan umur: ")

    # Kondisi berhenti
    if umur_input == "":
        break

    umur = int(umur_input)
    harga_tiket = 0.0 # Default harga

    # Logika menentukan harga tiket
    if umur <= 2:
        print("Gratis")
    elif umur <= 12:
        harga_tiket = 14.00
    elif umur >= 65:
        harga_tiket = 18.00
    else:
        harga_tiket = 23.00

    # Cetak harga hanya jika tidak gratis
    if harga_tiket > 0:
        print(f"Harga $ {harga_tiket:.2f}")

    # Tambahkan ke total dan cetak running total
    total_harga += harga_tiket
    print(f"Running total: {total_harga:.2f}")

# Proses Pembayaran
print(f"\nTotal Tagihan Anda: ${total_harga:.2f}")
uang_bayar = float(input("masukkan jumlah uang: "))
kembalian = uang_bayar - total_harga

# Cetak kembalian atau pesan uang kurang
if kembalian >= 0:
    print(f"Running kembalian: {kembalian:.2f}")
else:
    # 'kembalian' di sini nilainya negatif, jadi kita pakai -kembalian
    print(f"Maaf, uang Anda kurang ${-kembalian:.2f}.")
```
